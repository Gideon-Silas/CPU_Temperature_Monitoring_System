# Task: CPU TEMPERATURE MONITORING SYSTEM
A computer system records CPU temperature readings during operation.
## Given the following data:
### 45, 50, 55, 48, 52

```python
# 1. Store the values in a Python tuple

cpu_temps = (45, 50, 55, 48, 52)
print("CPU Temperatures:", cpu_temps)
```

```python
# 2. Print the first temperature reading

print("First temperature reading:", cpu_temps[0])
```

```python
# 3. Print the last temperature reading

print("Last temperature reading:", cpu_temps[-1])
```

```python
# 4. Find and print the difference between the highest and lowest temperatures

temp_difference = max(cpu_temps) - min(cpu_temps)
print("Highest temperature:", max(cpu_temps))
print("Lowest temperature:", min(cpu_temps))
print("Temperature difference:", temp_difference)
```

```python
# 5. Calculate and print the average CPU temperature

average_temp = sum(cpu_temps) / len(cpu_temps)
print("Average CPU temperature:", average_temp)
```

### Why a tuple is more suitable than a list for CPU temperature data

A tuple is better here because CPU temperature readings are recorded data that should not be changed or tampered with. Since tuples cannot be modified after they are created, they keep the data safe and accurate.
